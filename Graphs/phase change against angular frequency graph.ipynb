```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
high_damping_data = pd.read_excel("Labs phase shift high damping shift.xlsx")
low_damping_data = pd.read_excel("Labs phase shift low damping shift.xlsx")


#isolating the phase shift and the angular frequency data for high damping 

phase_shift_hd = high_damping_data['Phase shift']
angular_frequency = high_damping_data['Angular Frequency']



#isolating the phase shift and the angular frequency data for low damping 


phase_shift_ld = low_damping_data['Phase shift']
angular_frequency_1 = low_damping_data['Angular Frequency']

    
#plotting the graph for the high and low damping results on a graphs


plt.figure(figsize = (10,10))
plt.scatter(angular_frequency, phase_shift_hd, label = "high damping")
plt.errorbar(angular_frequency, phase_shift_hd, yerr = 0.06, xerr = 0.06, fmt ='o', linewidth =0.5, capsize = 2)



plt.scatter(angular_frequency, phase_shift_ld, label = "low damping")
plt.errorbar(angular_frequency, phase_shift_ld, yerr = 0.04, xerr = 0.06, fmt ='o', linewidth = 0.5, capsize = 2)


plt.xlabel('angular frequency ω (rad/s)')
plt.ylabel('phase shift (°)')
plt.title('phase shift against angular frequency across resonance')

plt.legend(["high damping, 1500Ω","low damping, 500Ω"])

plt.grid()
#plt.tight_layout()
plt.savefig('peak-to-peak voltage across capacitor against oscillator frequency graph.jpeg',dpi=2000 ,orientation = 'landscape')


                      
```
