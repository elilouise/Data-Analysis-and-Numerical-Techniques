# **Task 1**: Polar to cartesian

<div class="alert alert-success">

<h3>A function to convert between polar and cartesian coordinates</h3>
    
In the code cell below, create a (properly formatted and structured) function named <b><tt>convertpolar</tt></b> to convert from two-dimensional polar coordinates $r, \theta$ to cartesian coordinates $x,y$, using the equations 
$$ x = r \cos(\theta) \\
y = r \sin(\theta) $$
    This function will need to return <b>two</b> values (i.e. $x$ and $y$). To do this, use a statement like <tt>return x,y</tt> as the final line of the function. When you call the function, you will use a statement like (for example): 

<center><tt> x,y = convertpolar(r,theta)</tt>,</center> 
    
    
Remember to:
* Include an appropriate docstring
* Include other comments as necessary
* Name your function <tt>convertpolar</tt> (the checkpoint tests will not work if the name is different)
    


</div>

```python
def convertpolar(r, theta):
    '''
    
    Function converts two-dimensional polar coordinates  r, theta to cartesian coordinates x, y 
    inputs: r, theta (angle)
    outputs: x, y 
    
    '''
    # calculating cartesian x coordinate
    x = r * np.cos(theta)
    # calculating cartesian y coordinate
    y = r * np.sin(theta)
    
    #output x and y
    return x,y
```

# **Task 2**: Galilean spiral

<div class="alert alert-success">

<h3>Galilean spiral</h3>
    Now that your function is working, create another function called <tt>Gal_spiral</tt> which will calculate and plot the "Galilean spiral" 
$$r = \theta^2 - l$$
for the range $0 \leq \theta \leq 10\pi$ with $l = 0.1$. Your function should return $\theta$, $l$ and $r$ as the output (in that order).
    
<br>
<br>Some helpful hints: 
<ul>
   <li>Use an array function to create a numpy array of values for $\theta$</li> 
   <li>Don't forget to label the plot with axis labels and a title.</li>

</ul>
 </div>   
 
 

```python
def Gal_spiral():
    '''
    Function that calculates and generates a Galilean spiral plot
    
    inputs: N/A
    outputs: theta, l, r
    '''
    
    
    
    L = 0.1
    
    # creating the range of theta-value arrays
    theta = np.linspace(0,10*np.pi, 500) # 500 points between 0 and 10 pi
    
    #calculating the r (radius) values  
    r = theta**2 - L
    
    #converting polar coordinates (r,theta) to cartesian coordinates (x,y)
    x, y = convertpolar(r, theta)
    
    plt.figure() # create a new figure window
    
    plt.title('Galilean spiral') # create title for plot
    plt.xlabel('x') # create label for x-axis
    plt.ylabel('y') # create label for y-axis

    plt.plot(x,y) #plotting the Galilean spiral
    
    return theta, L, r
    
    
    
    
```

# **Task 3**: Spiral in a loop

<div class="alert alert-success">
See how the spiral changes when $l$ is varied: In the code cell below, create a function called <tt> loop_spiral</tt> which uses a loop structure to plot the spiral for <b>five equally spaced values of $l$ starting at 0 and ending at 10</b>, for $\theta$ in the range $0 \leq \theta \leq 2\pi$. Your function should plot the spirals and return $\theta$, $l$ and $r$ as an output.

<br>
<br>Some helpful hints:
<ul>
    <li>You can reuse most of the code from your previous function <tt>Gal_spiral</tt> </li>
    <li>It's best to create an array for the $l$-values. What form of loop is most appropriate for this?</li> 
   <li>
       Remember you'll need a legend for any plot with more than one line/dataset. One way of labelling the different lines would be to use a command such as 
       
       plt.plot(x,y, label=f"$l = {< variable >:3.1f}$".format()

```python
def loop_spiral():
    
    L = np.linspace(0,10,5)
    # creating the range of theta-value arrays
    theta = np.linspace(0,2*np.pi, 500) # 500 points between 0 and 2 pi
    r_list = []
        
    
    for i in range(len(L)):
        
        # creating the range of theta-value arrays
        #theta = np.linspace(0,2*np.pi, 500) # 500 points between 0 and 2 pi
        
        #calculating the r (radius) values for different values of L 
        r = theta**2 - L[i]
        r_list.append(r)
        
        #converting polar coordinates (r,theta) to cartesian coordinates (x,y)
        x, y = convertpolar(r, theta)
        
        plt.plot(x,y, label=L[i]) #plotting the Galilean spiral
        plt.title('Galilean spiral') # create title for plot
        plt.xlabel('x') # create label for x-axis
        plt.ylabel('y') # create label for y-axis
        
    
    
    
    
    
    plt.legend() #creating the legend
    
    
    return theta, L, np.array(r_list)
    
    
```
